**Zadanie 1**   
RozwiƒÖzanie problemu liniowego pozwalajƒÖcego na wyznaczenie warunk√≥w do uzyskania jak najwiƒôkszych zarobk√≥w firmy.





```
from pulp import *

prob = LpProblem("Sprzedaz_okien",LpMaximize)

x1 = LpVariable("aluminiowe",0,None)
x2 = LpVariable("drewniane",0,None)

prob += 3*x1 + 5*x2
prob += x1 <= 4
prob += 2*x2 <= 12
prob += 3*x1 + 2*x2 <= 18

prob.solve()
okna = []
nazwa = []
for k in prob.variables():
  nazwa.append(str(k.name))
for k in range(2):
  okna.append(str(value(prob.variables()[k])))
  print(nazwa[k]," = ",okna[k])

print("Maksymalny zarobek : ",value(prob.objective)*1000,"$")

```

**Zadanie 2**
Rozbudowanie programu o mo≈ºliwo≈õƒá zmiany parametr√≥w.

Do realizacji tego zadania wykorzystane zosta≈Çy suwaki.

```
from pulp import *
import pandas as pd
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
import ipywidgets as widgets

style = {'description_width': 'initial'}
cena_alu = FloatSlider(min=0.1,max = 5, value= 3, description="cena", style= style )
cena_dre = FloatSlider(min=0.1,max = 10, value= 5, description="cen_dre", style= style )
czas_f1 = IntSlider(min = 1, max = 10, value= 4, description="cz_f1",style= style)
czas_f2 = IntSlider(min = 2, max = 15, value= 12, description="cz_f2",style= style)
czas_f3 = IntSlider(min = 3, max = 20, value= 18, description="cz_f3",style= style)


def Produkcja(cena_aluminiowe = 3, 
              cena_drewniane = 5, 
              cz_f1 = 4, 
              cz_f2 = 12, 
              cz_f3 = 18,):
   
    prob = LpProblem("produkcja",LpMaximize) 

    x1=LpVariable("aluminowe", 0)
    x2=LpVariable("drewniane", 0)

    prob += cena_aluminiowe*x1 + cena_drewniane*x2, "Zysk"

    prob += x1 <= cz_f1, "Fabryka 1"
    prob += 2*x2 <= cz_f2, "Fabryka 2"
    prob += 3*x1 + 2*x2 <= cz_f3, "Fabryka 3"

    prob.solve()
    for v in prob.variables():
        print("{} = {} ".format(v.name, round(v.varValue)))
    print("Zarobek = {} z≈Ç".format(round(value(prob.objective)*1000)))

    shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob.constraints.items()]
    print(pd.DataFrame(shadows))
    
interact(
    Produkcja,
    cena_aluminiowe = cena_alu,
    cena_drewniane  = cena_dre,
    cz_f1 = czas_f1,
    cz_f2 = czas_f2,
    cz_f3 = czas_f3,
    )
   
```
