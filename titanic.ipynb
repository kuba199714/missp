```
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```
train_data = pd.read_csv("../content/train.csv")
train_data.head()


test_data = pd.read_csv("../content/test.csv")
test_data.head()

women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)

import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score ,confusion_matrix
pd.set_option("display.max_rows", None, "display.max_columns", None)
df = pd.read_csv('../content/train.csv')
df = df.drop(['PassengerId','Name','Ticket','Fare','Cabin'], axis=1)

df['Sex']=df['Sex'].replace('male',0)
df['Sex']=df['Sex'].replace('female',1)
df['Embarked']=df['Embarked'].replace('S',1)
df['Embarked']=df['Embarked'].replace('C',2)
df['Embarked']=df['Embarked'].replace('Q',3)

df['Age']=df['Age'].fillna(0)
df['Sex']=df['Sex'].fillna(0)
df['Embarked']=df['Embarked'].fillna(0)


y = df['Survived']
x = df.drop(['Survived'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.25,random_state=27)

# print(x_train)
# print(x_test)
# print(y_train)
# print(y_test)

clf = MLPClassifier(hidden_layer_sizes=(100), max_iter=500, alpha=0.0001,solver='lbfgs', verbose=10,  random_state=21,tol=0.000000001)

clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
#print(y_pred[:10])
#print(y_test[:10])

print('Accuracy  :',accuracy_score(y_test, y_pred))
# cm = confusion_matrix(y_test, y_pred)
# cm

# sns.heatmap(cm, center=True)
# plt.show()

df_test = pd.read_csv('../content/test.csv')
df_test = df_test.drop(['PassengerId','Name','Ticket','Fare','Cabin'], axis=1)

df_test['Sex']=df_test['Sex'].replace('male',0)
df_test['Sex']=df_test['Sex'].replace('female',1)
df_test['Embarked']=df_test['Embarked'].replace('S',1)
df_test['Embarked']=df_test['Embarked'].replace('C',2)
df_test['Embarked']=df_test['Embarked'].replace('Q',3)

df_test['Age']=df_test['Age'].fillna(0)
df_test['Sex']=df_test['Sex'].fillna(0)
df_test['Embarked']=df_test['Embarked'].fillna(0)

# print(df_test)

y_test=clf.predict((df_test))
# print(y_test[:10])
y_test=pd.DataFrame(data=y_test,columns=['Survived'])
y_test=pd.DataFrame(y_test)

 
y_test.to_csv('newsubmission.csv', index=False)
print(y_test)
```
